```python
    import requests
    import pandas as pd
    import json
    from pyspark.sql.functions import *
```

```python
%sql
use catalog practice;
use schema api;
```

```python
url = "https://jsonplaceholder.typicode.com/users"
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
df = pd.json_normalize(data)
df.columns = ['id', 'name', 'username', 'email', 'phone', 'website', 'street',
       'suite', 'city', 'zipcode', 'lat',
       'lng', 'lat.name', 'catchPhrase', 'bs']
```

```python
df1 = spark.createDataFrame(df)
df1.write.mode('overwrite').format('parquet').save('/Volumes/practice/api/api_data')
```

```python
df2 = spark.read.format('parquet').load('/Volumes/practice/api/api_data')
df2.display()
```
